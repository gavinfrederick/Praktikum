```python
perpustakaan = {}

def tambah_buku():
    print("\n=== TAMBAH BUKU BARU ===")
    id_buku = input("Masukkan ID Buku: ")

    if id_buku in perpustakaan:
        print("Buku dengan ID tersebut sudah ada di sistem!")
        return

    judul = input("Masukkan Judul Buku: ")
    penulis = input("Masukkan Nama Penulis: ")
    tahun = input("Masukkan Tahun Terbit: ")
    tanggal_pinjam = input("Masukkan Tanggal Terakhir Dipinjam (dd-mm-yyyy / - jika belum): ")
    status = input("Masukkan Status (Tersedia/Tidak): ")

    # Menambahkan data ke dictionary (nested dictionary)
    perpustakaan[id_buku] = {
        "JUDUL BUKU": judul,
        "PENULIS": penulis,
        "TAHUN TERBIT": tahun,
        "TANGGAL TERAKHIR PINJAM": tanggal_pinjam,
        "STATUS": status
    }

    print("Buku berhasil ditambahkan!\n")


def tampilkan_buku():
    print("\n=== DAFTAR BUKU DALAM SISTEM ===")
    if not perpustakaan:
        print("Belum ada data buku yang tersimpan.")
        return

    for id_buku, data in perpustakaan.items():
        print(f"\nID BUKU         : {id_buku}")
        print(f"JUDUL BUKU      : {data['JUDUL BUKU']}")
        print(f"PENULIS         : {data['PENULIS']}")
        print(f"TAHUN TERBIT    : {data['TAHUN TERBIT']}")
        print(f"TERAKHIR PINJAM : {data['TANGGAL TERAKHIR PINJAM']}")
        print(f"STATUS          : {data['STATUS']}")
        print("-" * 40)


# Menu utama
while True:
    print("""
========= SISTEM MANAJEMEN PERPUSTAKAAN =========
1. Tambah Buku Baru
2. Tampilkan Semua Buku
3. Keluar
===============================================
    """)
    pilihan = input("Pilih menu (1/2/3): ")

    if pilihan == "1":
        tambah_buku()
    elif pilihan == "2":
        tampilkan_buku()
    elif pilihan == "3":
        print("Terima kasih telah menggunakan sistem ini!")
        break
    else:
        print("Pilihan tidak valid, coba lagi.\n")
```
