Buatlah program dengan input dan fungsi untuk menghitung:

A. Luas persegi panjang

B. Volume balok

C. Luas lingkaran

D. Volume bola

```python
# a. Luas persegi panjang
def hitungluas_pp(p, l):
  return p * l

# b. Volume persegi panjang
def hitungvolume_pp(p, l, t):
  return p * l * t

# c. Luas lingkaran
def luas_lingkaran(r):
  pi = 3.14
  return pi * r * r

# d. Volume bola
def volume_bola(r):
  pi = 3.14
  return (4/3) * pi * r * r * r

# Memilih operasi
print("<=== Mau menghitung apa? ===>")
print("1. Luas persegi panjang")
print("2. Volume balok")
print("3. Luas lingkaran")
print("4. Volume bola\n")

pilihan = int(input("Masukan pilihan anda: "))

if pilihan == 1:
  p = int(input("Masukan panjang: "))
  l = int(input("Masukan lebar: "))
  print("\nLuas persegi panjang: ", hitungluas_pp(p, l))

elif pilihan == 2:
  p = int(input("Masukan panjang: "))
  l = int(input("Masukan lebar: "))
  t = int(input("Masukan tinggi: "))
  print("\nVolume balok:", hitungvolume_pp(p, l, t))

elif pilihan == 3:
  r = float(input("Masukan jari jari: "))
  print("\nLuas lingkaran: ", luas_lingkaran(r))

else:
   r = float(input("Masukan jari jari: "))
   print("\nVolume Bola: ",
   volume_bola(r))
```
