```python
# Dictionary
print("A. Buat dictionary keranjang dengan data: 'Apel' : 5000, 'Pisang' : 3000, 'Jeruk' : 7000, 'Anggur' : 15000")
keranjang = {
    'Apel' : 5000,
    'Pisang' : 3000,
    'Jeruk' : 7000,
    'Anggur' : 15000,
}
print(keranjang, "\n")

print("B. Gunakan Perulangan untuk menghitung total harga semua barang")
print("\nC. Tampilkan setiap buah dan harganya\n")
print("D. Harga total belanja\n")

total_harga = 0
for buah, harga in keranjang.items():
  print(f"Buah {buah} Harganya \t: {harga}")
  print(f"Total harga sementara \t: {harga + total_harga} \n")
  total_harga += harga
print(f"\nTotal harga buah \t: {total_harga}")
```
