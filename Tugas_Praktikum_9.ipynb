1. Diberikan list berikut:

buah = ['apel', 'mangga', 'jeruk', 'anggur', 'pisang']

a. Tambahkan 'semangka' di akhir list

b. Sisipkan 'durian' di antara 'jeruk' dan 'anggur'

c. Hapus 'mangga' dari list

d. Ubah 'pisang' menjadi 'nanas'

e. Tampilkan 3 buah pertama

```python
buah = ["Apel", "Mangga", "Jeruk", "Anggur", "Pisang"]
print(buah)
print(f"Nama buah : {buah}\n")

# a. menambah 'semangka' di akhir list
buah.append("Semangka")
print(f"A. Menambah semangka diakhir : {buah}\n")

# b. Sisipkan 'durian' di antara 'jeruk' dan 'anggur'
buah.insert(3, "Durian")
print(f"B. Menyisipkan buah durian : {buah}\n")

# c. Hapus 'mangga' dari list
buah.remove("Mangga")
print(f"C. Menghapus buah mangga : {buah}\n")

# d. Ubah 'pisang' menjadi 'nanas'
buah [4] = "Nanas"
print(f"D. Mengganti pisang menjadi nanas : {buah}\n")

# e. Tampilkan 3 buah pertama
buah_khusus = buah[0], buah[1], buah[2]
print(f"E. 3 buah pertama {buah_khusus}")
```

2. Diberikan list berikut:

angka = [45, 12, 78, 23, 56, 89, 34]

a. Urutkan list secara ascending

b. Urutkan list secara descending

```python
angka = [45,12,78,23,56,89,34]
print(angka)

# a. Urutkan list secara ascending
angka.sort()
print(f"A. Angka ascending : {angka}\n")

# b. Urutkan list secara descending
angka.reverse()
print(f"B. Angka descending : {angka}")
```
